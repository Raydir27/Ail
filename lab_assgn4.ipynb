```python
#Q1
import sys
import copy

def heuristic_count(state):

  misplaced = 0
  for i in range(3):
    for j in range(3):
      if state[i][j] != fin_st[i][j]:
        misplaced += 1
  return misplaced

def move(state, direction):
  #Move the empty tile in the given direction

  new_state = copy.deepcopy(state)
  empty_row, empty_col = find_empty(new_state)
  if direction == "up" and empty_row > 0:
    new_state[empty_row][empty_col], new_state[empty_row - 1][empty_col] = new_state[empty_row - 1][empty_col], new_state[empty_row][empty_col]
  elif direction == "down" and empty_row < 2:
    new_state[empty_row][empty_col], new_state[empty_row + 1][empty_col] = new_state[empty_row + 1][empty_col], new_state[empty_row][empty_col]
  elif direction == "left" and empty_col > 0:
    new_state[empty_row][empty_col], new_state[empty_row][empty_col - 1] = new_state[empty_row][empty_col - 1], new_state[empty_row][empty_col]
  elif direction == "right" and empty_col < 2:
    new_state[empty_row][empty_col], new_state[empty_row][empty_col + 1] = new_state[empty_row][empty_col + 1], new_state[empty_row][empty_col]
  return new_state

def find_empty(state):

  for i in range(3):
    for j in range(3):
      if state[i][j] == 0:
        return i, j

def best_first_search():

  front = [(in_st, 0)] #Queue of states
  explored = []  # Use a list instead of a set
  while front:
    # Sort front(lowest first)
    front.sort(key=lambda x: heuristic_count(x[0]))
    current_state, cost = front.pop(0)# Cost = number of moves made

    # Check if already explored (iterate through list)
    if current_state not in explored:
      explored.append(current_state)

      # Check if goal state reached
      if current_state == fin_st:
        print("Goal reached! Steps:", cost)
        return


      for direction in ["up", "down", "left", "right"]:
        new_state = move(current_state, direction)
        new_cost = cost + 1
        front.append((new_state, new_cost))

  print("No solution found.")

in_st = [[2, 0, 3], [1, 8, 4], [7, 6, 5]]
fin_st = [[1, 2, 3], [8, 0, 4], [7, 6, 5]]

best_first_search()
```

```python
#Q3
import heapq

initial = [[2, 0, 3], [1, 8, 4], [7, 6, 5]]
goal = [[1, 2, 3], [8, 0, 4], [7, 6, 5]]

def heuristic_misplaced(state):
    misplaced = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] != 0 and state[i][j] != goal[i][j]:
                misplaced += 1
    return misplaced

def generate_neighbors(current_state):
    neighbors = []
    moves = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    x, y = get_blank_index(current_state)
    for dx, dy in moves:
        nx, ny = x + dx, y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = [row[:] for row in current_state]
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            neighbors.append(new_state)
    return neighbors

def get_blank_index(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return (i, j)

def a_star_search(start):
    open_list = []
    closed_set = set()
    heapq.heappush(open_list, (heuristic_misplaced(start), 0, start))

    while open_list:
        _, cost, current_state = heapq.heappop(open_list)

        if current_state == goal:
            return current_state, cost

        closed_set.add(tuple(map(tuple, current_state)))

        for neighbor in generate_neighbors(current_state):
            if tuple(map(tuple, neighbor)) not in closed_set:
                new_cost = cost + 1
                heapq.heappush(open_list, (new_cost + heuristic_misplaced(neighbor), new_cost, neighbor))

    if final_state:
        print('Final State:', final_state)
        print('Cost:', final_cost)
    else:
        print('No solution found.')

start = initial
a_star_search(start)
```
