8 Bit Puzzle

Constraints/Rules:



*  up > 0
*  down < 2
*  left > 2
*  right < 2

search,define states,

```python
from collections import deque

#the state of the puzzle
class PuzzleState:
    def __init__(self, puzzle, moves=0, previous=None):
        self.puzzle = puzzle
        self.moves = moves
        self.previous = previous

    # Get the blank tile position
    def get_blank_position(self):
        for i in range(3):
            for j in range(3):
                if self.puzzle[i][j] == 0:
                    return i, j

    # Get possible moves
    def get_possible_moves(self):
        i, j = self.get_blank_position()
        moves = []
        if i > 0:
            moves.append((-1, 0))
        if i < 2:
            moves.append((1, 0))
        if j > 0:
            moves.append((0, -1))
        if j < 2:
            moves.append((0, 1))
        return moves

    # Generate new states
    def generate_states(self):
        possible_moves = self.get_possible_moves()
        states = []
        i, j = self.get_blank_position()
        for move in possible_moves:
            x, y = move
            new_puzzle = [row[:] for row in self.puzzle]
            new_puzzle[i][j], new_puzzle[i+x][j+y] = new_puzzle[i+x][j+y], new_puzzle[i][j]
            states.append(PuzzleState(new_puzzle, self.moves + 1, self))
        return states

    # Check if the puzzle is solved
    def is_solved(self):
        return self.puzzle == [[0, 1, 2], [3, 4, 5], [6, 7, 8]]

    # Print the puzzle
    def print_puzzle(self):
        for row in self.puzzle:
            print(row)
        print()

# Breadth-First Search
def bfs(initial_state):
    frontier = deque([initial_state])
    explored = set()
    index = 0

    while index < len(frontier):
        state = frontier[index]
        index += 1

        if state.is_solved():
            return state

        explored.add(str(state.puzzle))
        for neighbor in state.generate_states():
            if str(neighbor.puzzle) not in explored:
                frontier.append(neighbor)
                explored.add(str(neighbor.puzzle))

    return None


# Define the initial puzzle configuration
initial_puzzle = [
    [1, 2, 3],
    [0, 4, 6],
    [7, 5, 8]
]

# Create the initial state object
initial_state = PuzzleState(initial_puzzle)

# Solve the puzzle using Breadth-First Search
solution = bfs(initial_state)

# Print the solution
if solution:
    print("Solution found in", solution.moves, "moves:")
    while solution:
        solution.print_puzzle()
        solution = solution.previous
else:
    print("No solution found.")
```

Traveling Saleman
Problem

```python
from sys import maxsize
from itertools import permutations

V = 4

# Implementation of traveling Salesman Problem
def travellingSalesmanProblem(graph, s):

  # Store all vertices apart from source vertex
  vertex = []
  for i in range(V):
    if i != s:
      vertex.append(i)

  # Store minimum weight and path
  min_path = maxsize
  min_path_list = []

  # Loop through all permutations of the remaining vertices
  for i in permutations(vertex):

    # Store current path weight (cost) and path list
    current_pathweight = 0
    current_path = [s]  # Start with the starting vertex

    # Calculate current path weight and build path list
    k = s
    for j in i:
      current_pathweight += graph[k][j]
      current_path.append(j)
      k = j
    current_pathweight += graph[k][s]  # Add cost to return to starting vertex
    current_path.append(s)  # Add starting vertex back to the path

    # Update minimum weight and path if current path is shorter
    if current_pathweight < min_path:
      min_path = current_pathweight
      min_path_list = current_path.copy()  # Avoid modifying original list

  return min_path, min_path_list

# Sample adjacency matrix representation of graph
graph = [[0, 10, 15, 20],
        [10, 0, 35, 25],
        [15, 35, 0, 30],
        [20, 25, 30, 0]]
s = 0
min_cost, path = travellingSalesmanProblem(graph, s)
print(f"Minimum cost: {min_cost}")
print(f"Path: {' -> '.join(str(x) for x in path)}")
```
