2-sat simulated annealing

```python
import math
import random  # Import the random module

def gen_problem(soln):
  a, b, c, d = soln
  clause_1 = not a or d
  clause_2 = c or b
  clause_3 = not c or not d
  clause_4 = not d or not b
  clause_5 = not a or not d
  return [clause_1, clause_2, clause_3, clause_4, clause_5]

def movegen(soln):
  soln_list = list(soln)
  random_index = random.randint(0, len(soln_list) - 1)
  soln_list[random_index] = "1" if soln_list[random_index] == "0" else "0"
  return soln_list

def energy(clauses):
  # Evaluate all clauses and return True if all are satisfied, False otherwise
  for clause in clauses:
    if not eval(str(clause)):  # Convert clause to string before evaluation
      return False
  return True


def simulated_annealing(initial_soln, initial_temp, cooling_rate, iterations):
  current_soln = initial_soln
  current_temp = initial_temp
  for i in range(iterations):
    if current_temp <= 0:
      break
    new_soln = movegen(current_soln)
    delta_E = energy(gen_problem(new_soln)) - energy(gen_problem(current_soln))
    probability = math.exp(-delta_E / current_temp)
    if probability > random.random():  # Metropolis acceptance criteria
      current_soln = new_soln
    current_temp *= cooling_rate
  return current_soln


initial_soln = [0, 1, 0, 1]
solution = simulated_annealing(initial_soln, 100, 0.95, 1000)
if energy(gen_problem(solution)):
  print("Solution found:", solution)
else:
  print("No solution found within the limit")
```
